```python
import os
import pandas as pd

directory = "/Users/vincewu/Bees-Project"
processed_folder = "Fully Processed"
processed_file = 'forager_data.csv'

processed_csv = pd.read_csv(os.path.join(directory, processed_folder, processed_file), engine='python', header=None)

headers=processed_csv.loc[0].tolist()
headers[0]='Numbers'

processed_csv.columns=headers
processed_csv=processed_csv.drop(labels=0, axis=0)
processed_csv.index-=1
processed_csv

```

```python
# Average total trip time per bee
# Average rest duration per bee
# Average trip duration for all bees
# Average rest duration for all bees bee
# Average number of trips taken
# Peak hour of activity (hour with most trips taken)
# Average Start hour
# Average End hour

min_day=1000
min_date=''
max_day=0
max_date=''
for i in range(len(processed_csv)):
    temp=int(processed_csv['Day'][i])
    if temp>max_day: 
        max_day=temp
        max_date=processed_csv['Date'][i]
    if temp<min_day: 
        min_day=temp
        min_date=processed_csv['Date'][i]

total_range=max_day-(min_day-1)
trip_times_per_bee=[0 for days in range(total_range)]
rest_durations_per_bee=[0 for days in range(total_range)]
trip_times_overall=[0 for days in range(total_range)]
rest_durations_overall=[0 for days in range(total_range)]
trips_taken = [0 for days in range(total_range)]
trips_each_hour = [list(0 for hours in range(24)) for days in range(total_range)]
start_hour = [24 for days in range(total_range)]
end_hour = [0 for days in range(total_range)]
total_bees = [0 for days in range(total_range)]


# print(min_day, min_date, max_day, max_date)
# print(trips_each_hour)

for i in range(len(processed_csv)):
    trip_amount=int(processed_csv['Trip Amount'][i])

    if trip_amount>0:
        day=int(processed_csv['Day'][i])-min_day
        # print(day)
        total_bees[day]+=1

        trip_duration=float(processed_csv['Trip Average Duration (Hours)'][i])
        trip_times_per_bee[day]+=trip_duration*trip_amount

        trip_times_overall[day]+=trip_duration*trip_amount
        trips_taken[day]+=trip_amount

        rest_duration = float(processed_csv['Rest Average Duration (Hours)'][i])
        rest_durations_per_bee[day]+=rest_duration*(trip_amount-1)
        rest_durations_overall[day]+=rest_duration*(trip_amount-1)

        start_times = processed_csv['All Start Times (Hours)'][i]
        start_times=start_times.split(', ')
        start_times[0]=start_times[0][1:]
        start_times[-1]=start_times[-1][:-1]

        end_times = processed_csv['All End Times (Hours)'][i]
        end_times=end_times.split(', ')
        end_times[0]=end_times[0][1:]
        end_times[-1]=end_times[-1][:-1]

        # if int(float(start_times[0]))<5:
        #     print(processed_csv['UID'][i], day+1, int(float(start_times[0])), sep='\n')

        if int(float(start_times[0])) < start_hour[day]:
            start_hour[day]=int(float(start_times[0]))
        if int(float(end_times[-1])) > end_hour[day]:
            end_hour[day]=int(float(end_times[-1]))

        for i in range(len(start_times)):
            start_time=int(float(start_times[i]))
            end_time=int(float(end_times[i]))
            for hour in range(start_time,end_time+1):
                trips_each_hour[day][hour]+=1


print(trip_times_per_bee, rest_durations_per_bee,
      trip_times_overall, rest_durations_overall, trips_taken,
      trips_each_hour, start_hour, end_hour, total_bees, sep='\n \n')
```

```python
import math

total_range=max_day-(min_day-1)
#average minutes in that hour spent outside per bee that was outside for that hour
hourly_trip_times_per_active_bee = [
    [0 for hours in range(24)] for days in range(total_range)]
#average minutes in that hour spent outside per bee that was outside for that day
hourly_trip_times_per_overall_bee = [
    [0 for hours in range(24)] for days in range(total_range)]
#average duration of trips taken during that hour
hourly_trip_times_overall=[[0 for hours in range(24)] for days in range(total_range)]
#rest_durations_overall=[[0 for hours in range(24)] for days in range(total_range)]
#percent of total trips started during that hour
hourly_trips_started = [[0 for hours in range(24)] for days in range(total_range)]
#percent of total number of trips started or ocurring during that hour
hourly_trips_occuring = [[0 for hours in range(24)] for days in range(total_range)]
#percent of total number of individual bees active that day
#who spent some time outside during that hour
hourly_bee_activity = [[0 for hours in range(24)] for days in range(total_range)]

# print(min_day, min_date, max_day, max_date)
# print(trips_each_hour)

for i in range(len(processed_csv)):
    trip_amount=int(processed_csv['Trip Amount'][i])

    if trip_amount>0:
        day=int(processed_csv['Day'][i])-min_day
        # print(day)
        
        trips=processed_csv['Trip Start and End Times (Hours)'][i]
        trips=trips.replace('[', '').replace(']', '').split(', ')

        rounded_trips=trips.copy()

        for j in range(len(rounded_trips)):
            for k in range(len(rounded_trips[j])):
                if rounded_trips[j][k]=='.':
                    index=k
                    break
            rounded_trips[j]=int(rounded_trips[j][:index])

        for j in range(len(trips)):
            trips[j]=float(trips[j])

        for j in range(len(trips)):
            if j%2==1:
                trip_start_hour=rounded_trips[j-1]
                trip_end_hour=rounded_trips[j]

                # if trip_start_hour<3 and day==0:
                #     print(i)
                
                for hour in range(trip_start_hour+1, trip_end_hour):
                    hourly_trip_times_per_active_bee[day][hour]+=1
                    hourly_trip_times_per_overall_bee[day][hour]+=1
                    hourly_trip_times_overall[day][hour]+=1

                if trip_start_hour==trip_end_hour:  
                    hourly_trip_times_per_active_bee[day][trip_start_hour]+=trips[j]-trips[j-1]
                    hourly_trip_times_per_overall_bee[day][trip_start_hour]+=trips[j]-trips[j-1]
                    hourly_trip_times_overall[day][trip_start_hour]+=trips[j]-trips[j-1]
                else:
                    hourly_trip_times_per_active_bee[day][trip_start_hour] += 1-trips[j-1]%1
                    hourly_trip_times_per_active_bee[day][trip_end_hour] += trips[j]%1
                    hourly_trip_times_per_overall_bee[day][trip_start_hour] += 1-trips[j-1]%1
                    hourly_trip_times_per_overall_bee[day][trip_end_hour] += trips[j]%1
                    hourly_trip_times_overall[day][trip_start_hour] += 1-trips[j-1]%1
                    hourly_trip_times_overall[day][trip_end_hour] += trips[j] % 1

        outside = [False for hour in range(24)]
        for j in range(len(rounded_trips)):
            if j%2==0:
                hourly_trips_started[day][rounded_trips[j]] += 1
            elif j%2==1:
                for k in range(rounded_trips[j-1], rounded_trips[j]+1):
                    hourly_trips_occuring[day][k] += 1
                    outside[k]=True
                

        for hour in range(len(outside)):
            if outside[hour]:
                hourly_bee_activity[day][hour] += 1
        # if outside[0] or outside[1] or outside[2]:
        #     print(i)
    

print(hourly_trip_times_per_active_bee[0], hourly_trip_times_per_overall_bee[0],
hourly_trip_times_overall[0], 
hourly_trips_started[0],
hourly_trips_occuring[0],
hourly_bee_activity[0], sep='\n \n')
```

```python
for day in range(total_range):
    for hour in range(24):
        if hourly_bee_activity[day][hour]>0:
            hourly_trip_times_per_active_bee[day][hour] /= hourly_bee_activity[day][hour]
        if total_bees[day]>0:
            hourly_trip_times_per_overall_bee[day][hour]/=total_bees[day]
        if hourly_trips_occuring[day][hour]>0:
            hourly_trip_times_overall[day][hour]/=hourly_trips_occuring[day][hour]
        if trips_taken[day]>0:
            hourly_trips_started[day][hour]/=trips_taken[day]
            hourly_trips_occuring[day][hour]/=trips_taken[day]
        if total_bees[day]>0:
            hourly_bee_activity[day][hour]/=total_bees[day]

print(hourly_trip_times_per_active_bee[0], hourly_trip_times_per_overall_bee[0],
hourly_trip_times_overall[0], 
hourly_trips_started[0],
hourly_trips_occuring[0],
hourly_bee_activity[0], sep='\n \n')

hourly_bee_data=[hourly_trip_times_per_active_bee, hourly_trip_times_per_overall_bee,
hourly_trip_times_overall, hourly_trips_started, hourly_trips_occuring,
hourly_bee_activity]
```
